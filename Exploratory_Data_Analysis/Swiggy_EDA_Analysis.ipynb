
### Top 10 resturants based on total and average ratings by the customers.

```python
n=10 # top 10 resturant based on ratings
total_rating_Swiggy=df.nlargest(n,"Avg ratings")
bar_color="blue"
vis_total_ratings=px.bar(total_rating_Swiggy,x="Restaurant",y="Total ratings",title=f"Top {n} Restaurant by Total Ratings",labels={"Total ratings":"Total Ratings"},color_discrete_sequence=[bar_color])
vis_total_ratings.show()
average_rating_swiggy=df.nlargest(n,"Avg ratings")
bar_color="red"
vis_average_rating=px.bar(average_rating_swiggy,x="Restaurant",y="Avg ratings",title=f"Top {n} Restaurant by Average Ratings",labels={"Avg ratings":"Average raing"},color_discrete_sequence=[bar_color])
vis_average_rating.show()
```


### Top 10 Cities Ordering Food from Swiggy


### Areas with Highest Order Counts for Each City

```python
city_area_order_counts = df.groupby(['City', 'Area']).size().reset_index(name='Order Count')
city_area_order_counts_sorted = city_area_order_counts.sort_values(by='Order Count', ascending=False)
fig = px.bar(
    city_area_order_counts_sorted,
    x='Area',
    y='Order Count',
    color='City',  # Use different colors for each city's bars
    title='Areas with Highest Order Counts for Each City',
    labels={'Order Count': 'Order Count', 'Area': 'Area'},
)
fig.show()
```


### Most Popular Food Types in Top 10 Cities

```python
city_food_counts = df.groupby(['City', 'Food type']).size().reset_index(name='Count')
popular_food_types = city_food_counts.groupby('City').apply(lambda x: x.loc[x['Count'].idxmax()])
top_10_cities = popular_food_types.nlargest(10, 'Count')
# Prepare the data for treemap
labels = top_10_cities['City'] + ' - ' + top_10_cities['Food type']
sizes = top_10_cities['Count']

# Create the treemap
plt.figure(figsize=(12, 6))
squarify.plot(sizes=sizes, label=labels, alpha=0.8)
plt.axis('off')
plt.title('Most Popular Food Types in Top 10 Cities')
plt.show()
```
